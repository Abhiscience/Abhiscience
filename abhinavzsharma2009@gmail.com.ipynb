__Data analysis pre-profiling observations::__

- The data set has 4 numerical columns.
- First we will check price and mileage column having zero values together.
- The data shows **price** of the cars having all values but **267 ZEROS** values. A used car on sale cant be given for free. These data needs to be changed. The ***price*** data shows high **kurtosis** and high **skewness**. The **mean** is quite high then **median** depicting the data is more skewed towards right. **Positively Skewed!**. *The zero values are being replaced by median values.* 
- The price is in $ Source :: Dictionary https://raw.githubusercontent.com/insaid2018/Term-1



```python
data[(data['price']==0)&(data['mileage']==0)]
# 45 Values where price and mileage are both zero. This wont give any analysis, hence dropping these 45 values
```

- Checking How many vehicles have price and mileage together crossing there standard deviation.

```python
data[(data['price']>=3*data['price'].std())&(data['mileage']>=3*data['mileage'].std())]
```

```python
data.drop(data[(data['price']==0)&(data['mileage']==0)].index,axis=0,inplace=True)
data.info()
```

- After *45* values are dropped from the common **price** & **mileage** having **0** values. **9531** rows are left.

- **Price** Column needs treatment, the zero values are to be replaced with median, since it is highly skewed and with high kurtiosis.

```python
# Using Replace function to replace zero with median of the price column.
data.price.replace(to_replace=0,value=data['price'].median(),inplace=True)
data.price.min()
```

__The Outliers in PRICE Column is too High! It may lead to the following__

- The data points after **3σ** **(99.7%)**  values are considered as outliers.
- All values above **3σ** are being saved as luxury car price segment.

```python
data_luxury=data[data['price']>data['price'].std()*3]
data_luxury
```

****Droping The Luxury Car Data Now from Main Data Frame****

- Using Drop Function on main data set DATA, the function will be executed later. Hence it is not performed right now.

```python
#data.drop(data[data['price']>3*data['price'].std()].index,inplace=True)
#data
```


__The data analysis on pre-profiling observations::__

 ***MILEAGE***
- The data shows **mileage** of the cars having all values but **343 ZERO** values after removing **45** values common to price and mileage. A used car on sale cant have zero mileage, else it would be a new car. These data needs to be changed. The ***mileage*** data shows high **kurtosis** and high **skewness**. The **mean** is quite high then **median** depicting the data is more skewed towards right. **Positively Skewed!**. *The zero values are being replaced by median values.* 
- As per dictionary the mileage is per 1000 kms. https://raw.githubusercontent.com/insaid2018/Term-1

```python
data.mileage.replace(to_replace=0,value=data['mileage'].median(),inplace=True)
data.mileage.min()
```

* Checking no of values lying beyound **3σ (99.7%)** in *Mileage* (which is per 1000 kms)**

```python
data_highrun_Cars=data[data['mileage']>3*data['mileage'].std()]
data_highrun_Cars.mileage.unique()
# 716 Cars Data Set has milage more than 3σ of mileage. We have classified it has high run vehicles. 
#Hence may be we can use it later in analysis.
```

- **The values of engV cannot exceed 10 Litrs, hence dividing all the values of engine more than 10 liters by 10**
- **The values of engV less than 

- **There are missing values in engV also, the analysis shows it to be right skewed as** **mean** **is more than **median**. We are replacing the Nan Values with median**

```python
data.engV=data.engV.apply(lambda x: x/10.0 if x>10.0 else x)
data.engV.fillna(data.engV.median(),inplace=True)
data.engV.unique()
```

```python
data.engV=data.engV.apply(lambda x: x*10.0 if x<0.8 else x)
data.engV.unique()
```
